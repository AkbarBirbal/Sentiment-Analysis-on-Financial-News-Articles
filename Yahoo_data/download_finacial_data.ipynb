```python
import urllib.request
import json
import csv
```

```python
fi = open('symols','r')
data = json.load(fi)
```

```python
burl = 'http://ichart.finance.yahoo.com/table.csv?s='
```

```python
val =[]
for keys,vals in data.items():
    val.append([keys,vals])
```

```python
val[0]
```

```python
output_path = "D:\\data"
write_path ="D:\\data\\updated"
```

```python
not_found=[]
for i in val:
    url = burl+i[1]
    output_path_company=output_path+'\\'+i[0]+'.csv'
    try:
        urllib.request.urlretrieve(url,output_path_company)
        with open(output_path_company,'r') as csvinput:
            with open(write_path+'\\'+i[0]+'.csv', 'w') as csvoutput:
                writer = csv.writer(csvoutput, lineterminator='\n')
                reader = csv.reader(csvinput)
                all = []
                row = next(reader)
                row.append('company')
                all.append(row)
                for row in reader:
                    row.append(i[0])
                    all.append(row)
                writer.writerows(all)  
    except:
        not_found.append(i)
        pass
    
```

```python
not_found
```

```python
import pickle 
pickle.dump(not_found, open('not_found_historcal_data.list', 'wb')) 
```

```python
mer = []
for i in val:
    if i not in not_found:
        mer.append(i)
```
